```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r_min, r_max = 2.5, 4.0  # Range of r values
iterations = 1000  # Number of iterations to discard for stability
last = 100  # Number of iterations to plot after discarding initial transient

# Number of r values to plot
r_values = np.linspace(r_min, r_max, 10000)
x = 1e-5 * np.ones(len(r_values))  # Initial population values (all start near zero)

# Set up the plot
plt.figure(figsize=(5, 3))
for i in range(iterations + last):
    x = r_values * x * (1 - x)  # Logistic map iteration
    if i >= iterations:  # Only plot the last few iterations
        plt.plot(r_values, x, ',k', alpha=0.25)  # Plot with a small marker

# Label the diagram
plt.title("Bifurcation Diagram of the Logistic Map")
plt.xlabel("$\mu$")  # x-label
plt.xlim(r_min, r_max)
plt.ylim(0, 1)

# Adjust layout to make sure all labels are visible
plt.tight_layout()

# Save the figure before showing it
plt.savefig("logistic_map_bifur.png")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

# Logistic map function
def logistic_map(mu, x0=0.5, n_iter=1000, n_skip=100):
    x = x0
    values = []
    for _ in range(n_iter):
        x = mu * x * (1 - x)
        if _ >= n_skip:  # Skip some initial iterations to allow stabilization
            values.append(x)
    return values

# Function to plot logistic map for a given mu
def plot_logistic_map(mu):
    x_values = logistic_map(mu)
    plt.figure(figsize=(5, 3))
    plt.plot(range(len(x_values)), x_values, ',k', lw=0.5)
    plt.title(f"Logistic Map for  = {mu}")
    plt.xlabel("Iteration")
    plt.ylabel("x")
    plt.grid(True)
    plt.show()

# Create an interactive widget for controlling mu
mu_slider = widgets.FloatSlider(min=2.5, max=4.0, step=0.01, value=3.5, description='')

# Connect the slider to the plot function
interact(plot_logistic_map, mu=mu_slider)
```
