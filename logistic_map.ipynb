```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r_min, r_max = 2.5, 4.0  # Range of r values
iterations = 1000  # Number of iterations to discard for stability
last = 100  # Number of iterations to plot after discarding initial transient

# Number of r values to plot
r_values = np.linspace(r_min, r_max, 10000)
x = 1e-5 * np.ones(len(r_values))  # Initial population values (all start near zero)

# Set up the plot
plt.figure(figsize=(10, 7))
for i in range(iterations + last):
    x = r_values * x * (1 - x)  # Logistic map iteration
    if i >= iterations:  # Only plot the last few iterations
        plt.plot(r_values, x, ',k', alpha=0.25)  # Plot with a small marker

# Label the diagram
plt.title("Bifurcation Diagram of the Logistic Map")
plt.xlabel("r")
plt.ylabel("Population")
plt.xlim(r_min, r_max)
plt.ylim(0, 1)

# Show the plot
plt.show()
```
